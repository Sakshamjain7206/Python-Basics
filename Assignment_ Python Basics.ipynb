2.What is an interpreter in Python?

Ans- An interpreter in Python is a program that executes Python code line by line. Unlike a compiler, which translates the entire source code into machine code before execution, an interpreter reads and executes the code directly, converting each line into a form the computer can execute at runtime.
This interpretation process is what allows Python to be cross-platform and easy to use in various environments without the need for compiling the code.

5.What is mutability in Python?

Ans- Mutability in Python refers to the ability of an object to be modified after it is created. In other words, a mutable object’s state or value can be changed, whereas an immutable object cannot be altered once it is assigned.

6.Why are lists mutable, but tuples are immutable?

10.What is the difference between implicit and explicit type casting?

Ans- The difference between implicit and explicit type casting in Python lies in how the conversion of data types is performed and who initiates it.

1. Implicit Type Casting (Automatic)
Definition: In implicit type casting, Python automatically converts a smaller data type to a larger one, usually when there is no risk of losing information or precision.
2. Explicit Type Casting (Manual)
Definition: In explicit type casting, the programmer manually converts one data type to another using built-in functions like int(), float(), str(), etc.

11.What is the purpose of conditional statements in Python?

Ans- The purpose of conditional statements in Python is to allow the program to make decisions and execute specific blocks of code based on certain conditions. They enable the program to respond differently depending on whether a condition is True or False, thus controlling the flow of execution.

Types of Conditional Statements in Python:
1. if statement: Executes a block of code if the condition is True.
2. elif statement (short for "else if"):Checks additional conditions if the previous if conditions are False.
3. else statement:Executes a block of code if all preceding if/elif conditions are False.

12.How does the elif statement work?

Ans-The elif (short for "else if") statement in Python is used to check multiple conditions sequentially after an initial if statement. It allows you to test several conditions in a chain, executing a block of code as soon as one of those conditions is True. If none of the conditions are true, the else block (if provided) will execute.


13.What is the difference between for and while loops?

Ans-The for and while loops in Python are both used for iteration, but they differ in how they control the loop’s execution.

1. for Loop
Definition: A for loop is used to iterate over a sequence (like a list, tuple, string, or range). It repeats the block of code for each item in the sequence.
2. while Loop
Definition: A while loop repeats a block of code as long as a given condition evaluates to True. It continues running until the condition becomes False.


14.Describe a scenario where a while loop is more suitable than a for loop?

Ans-A while loop is more suitable than a for loop when the number of iterations is not known in advance and you need to continue looping until a specific condition is met, which might depend on dynamic factors.

**Example Scenario: User Input Validation**
Imagine a scenario where you want to repeatedly prompt the user to enter a valid password until they enter one that meets certain criteria (e.g., at least 8 characters, contains both letters and numbers). Since you don't know how many attempts the user will need, a while loop is ideal.

```python
#6/7 Write a function to demonstrate the behavior of mutable and immutable arguments
# Function to demonstrate mutable and immutable arguments
def modify_arguments(mutable_arg, immutable_arg):
    print("Before modification:")
    print("Mutable argument:", mutable_arg)
    print("Immutable argument:", immutable_arg)

    # Modify the mutable argument (list)
    mutable_arg[0] = 100

    # Modify the immutable argument (integer)
    immutable_arg = 200

    print("\nAfter modification:")
    print("Mutable argument:", mutable_arg)  # The change is reflected
    print("Immutable argument:", immutable_arg)  # The change is local to the function

# Example usage
my_list = [1, 2, 3]  # Mutable argument
my_number = 10       # Immutable argument

modify_arguments(my_list, my_number)

# Outside the function
print("\nOutside the function:")
print("List after function call:", my_list)  # List will be modified
print("Number after function call:", my_number)  # Number will remain unchanged
```

```python
#8 Write a program to demonstrate the use of logical operators
# Example variables
x = 10
y = 5
z = 20

# Using the 'and' operator: Returns True only if both conditions are True
if x > y and z > x:
    print("Both conditions are True using 'and'.")

# Using the 'or' operator: Returns True if at least one condition is True
if x > y or x < z:
    print("At least one condition is True using 'or'.")

# Using the 'not' operator: Reverses the truth value of the condition
if not (x < y):
    print("The condition x < y is False, so 'not' makes it True.")

# Combining logical operators
if (x > y and z > x) or (y > z):
    print("Combined conditions using 'and' and 'or' are True.")
```

```python
#13 Write a Python program to find the sum of all even numbers between 1 and 50
# Initialize sum variable
even_sum = 0

# Loop through numbers from 1 to 50
for number in range(1, 51):
    if number % 2 == 0:  # Check if the number is even
        even_sum += number  # Add even number to the sum

# Print the sum of even numbers
print("The sum of all even numbers between 1 and 50 is:", even_sum)
```

```python
#14 Write a program to reverse a string using a while loop
# Input string
input_string = "Hello, World!"

# Initialize an empty string to store the reversed string
reversed_string = ""

# Initialize an index to point to the last character of the input string
index = len(input_string) - 1

# While loop to reverse the string
while index >= 0:
    reversed_string += input_string[index]  # Add the character to reversed_string
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_string)
```

```python
#15 Write a Python program to calculate the factorial of a number provided by the user using a while loop
# Get input from the user
number = int(input("Enter a number to calculate its factorial: "))

# Initialize variables
factorial = 1
counter = number

# While loop to calculate the factorial
while counter > 1:
    factorial *= counter  # Multiply current factorial by counter
    counter -= 1  # Decrement the counter

# Print the factorial result
print(f"The factorial of {number} is {factorial}")
```
